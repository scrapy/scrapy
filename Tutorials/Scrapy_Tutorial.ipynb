## Setup Project


```python
import scrapy
import os
```

```python
# Names
PROJ_NAME='example'
SPIDER_NAME='example_spider'
OUTPUT_FILE='output.csv'
```

```python
# Create a project and cd into it
cmd = f'scrapy startproject {PROJ_NAME}'
os.system(cmd)
os.chdir(f'{PROJ_NAME}')
```

## Create Spider

```python
# spider file
%%writefile /content/{PROJ_NAME}/{PROJ_NAME}/spiders/{SPIDER_NAME}.py
import scrapy 


class MySpider(scrapy.Spider):
  name= 'myspider'
  start_urls = ['https://www.zyte.com/blog/']
  
  def parse(self, response):
    for title in response.css('.oxy-post-title'):
      yield {'title': title.css('::text').get()}

    for next_page in response.css('a.next'):
      yield response.follow(next_page, self.parse)
```

## Crawl


### Print output

```python
import subprocess
from pprint import pprint

# replace 'myspider' with the the spider name
cmd = f'scrapy crawl myspider'

proc = subprocess.run(["scrapy", "crawl", "myspider"], capture_output=True)
pprint(proc.stdout)
pprint(proc.stderr)
```

### To file

```python
# replace 'myspider' with the name of the spider
cmd = f'scrapy crawl myspider -O {OUTPUT_FILE}'
os.system(cmd)
```
